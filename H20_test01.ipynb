# Trial of H20 AutoML
### Previous notebooks used various deep neural networks to perform predictions.

Note: Java needs to be installed for this to run. I ran this on a relatively new MacBook Pro, and needed to install Homebrew and java. Once those were installed, this code ran perfectly.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import h2o
from h2o.automl import H2OAutoML

# initialize h2o and load training data
h2o.init()
```

```python
# load data
data = h2o.import_file("data/train.csv")
# set target variable
data['Transported'] = data['Transported'].asfactor()

# split data into train and test
train, test = data.split_frame(ratios=[0.8], seed=48)
```

### Run H20 AutoML

```python
# Configure and run AutoML
# max_runtime_secs is the time limit; adjust as needed
automl = H2OAutoML(max_runtime_secs=3600, seed=1, project_name='kaggle_project')
automl.train(y='Transported', training_frame=train)


```

```python
# view the AutoML Leaderboard
lb = automl.leaderboard
print(lb)
```

```python
# examine the top model
model_id = lb[0, 'model_id']

model = h2o.get_model(model_id)

model.summary()

```

```python
# explore model parameters
model_params = model.params
print(model_params)

# get model performance
model_performance = model.model_performance()
print(model_performance)
```

```python
# run the model on the new test data
test_data = h2o.import_file("data/test.csv")
test_data_df = pd.read_csv("data/test.csv")

# save the PassengerId for submission
passenger_id = test_data_df['PassengerId']

# get the best model
model_best = automl.leader

# make predictions and convert to df
preds = model_best.predict(test_data)
preds_df = preds.as_data_frame()

# print the predictions
print(preds_df.head())
```

### It looks like H20 did not split the Cabin column in the 3 seperate columns.
- It seems to have done very well without that step.
- This notebook will continue processing the test data for submission. The next notebook will split the Cabin column prior to running AutoML


```python
submission = pd.DataFrame({'PassengerId': passenger_id, 'Transported': preds_df['predict']})

# save the submission file as csv
submission.to_csv('data/submission_autoML_01.csv', index=False)
```
